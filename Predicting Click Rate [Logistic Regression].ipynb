# Background
The dataset consists of data from a fictional internt company with fictional advertising data, indicating whether or not a particular internet user clicked on an Advertisement on a company website. 

# Objective
To create a model that will predict whether or not a user will click on an ad based based on user characteristics.

# Dataset Description
This data set contains the following features:

* Daily Time Spent on Site: consumer time on site in minutes
* Age: cutomer age in years
* Area Income: Avg. Income of geographical area of consumer
* Daily Internet Usage: Avg. minutes a day consumer is on the internet
* Ad Topic Line: Headline of the advertisement
* City: City of consumer
* Male: Whether or not consumer was male
* Country: Country of consumer
* Timestamp: Time at which consumer clicked on Ad or closed window
* Clicked on Ad: 0 or 1 indicated clicking on Ad

## Setting up the model

**Removing categorical feature**

```python
ad_data.drop(labels = ['Ad Topic Line', 'City', 'Country', 'Timestamp'], axis = 1, inplace = True)
```

```python
X = ad_data.drop('Clicked on Ad', axis = 1)
y = ad_data['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

## Predictions and Evaluations

```python
y_pred = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print('Classification Report')
print(classification_report(y_test, y_pred))
print('\n')
print('Confusion Matrix')
print(confusion_matrix(y_test, y_pred))
```

## Conclusion

**The Precision and Recall are calculated as follows**

![alt text](Confusion Matrix_1.png "Title")

**Precision** reveals how precise the model is. Meaning, out of those predicted positives, how many of them are actual positive. 

Out of the predicted 'Clicked on Ad', the model correctly predicted 93% of the values. This means that 7% of the values were false positives. In other words, 7% of the time, the model returned a 'Clicked on Ad' when the user did not click on the ad. 


**Recall** calculates how many of the Actual Positives the model captured through labeling it as Positive (True Positive).

Out of the predicted 'Clicked on Ad', the model correctly predicted 90% of the values. This means that 10% of the values were false negatives. In other words, 10% of the time, the model returned a didn't 'Click on Ad' when the user did click on the ad.
